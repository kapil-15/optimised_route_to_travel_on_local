```python
import camelot.io as camelot
import pandas as pd

tables = camelot.read_pdf(
    r"C:/Users/kapil/Downloads/central_line_.pdf",
    flavor='stream',
    pages="all",
	# strip_text="\n",
    # split_text=True,
    # copy_text=["h"]
	#_______________________________
	# process_background=True,  # Detect light borders
    row_tol=2,               # Tighten row detection
    edge_tol=750,
	#_______________________________
	table_areas=["50,800,800,0"]               #["50,700,550,100"] ## Left, Top, Right, Bottom
)

count = 0
for i in tables:
    print(i)
    count += 1
    df = i.df
    header_rows = df.loc[1:4]
    new_header = header_rows.apply(lambda x: '_'.join(x.astype(str)), axis=0)
    df.columns = new_header
    df = df.iloc[5:]
    df.reset_index(drop=True, inplace=True)
    # df.to_csv(f"../data/raw/Central_Line_{count}.csv")

```
